#### Check for nulls after dropping nulls from City and Zipcode

```python
df_2020_C_Z_dropped = df_2020.dropna(how = 'any', subset = ['City', 'ZipCode'])
df_2020_C_Z_dropped.isnull().sum()
```

#### Null Summation
2020 original length: 1207612 

2020 rows with null values: 9908 = 1197704 *

length of dropped nulls: 1197704

**Column Zipcode nulls include all Column City nulls*

#### Summation Calculation below

```python
1197704 + 9908
```

#### What the Summation Calculation means...
There are 9908 rows with null values and there are 9908 rows with null ZipCode values. Therefore of those 9908 rows with null Zipcode Values, there are 2589 rows with null City Values. The remainder (9908 - 2589 = 7319) rows with non-null city values and null Zipcode Values can be filled in based on Matching Latitudes and Longitudes.

Some further notes:
Googling the lat and long of entry number 164360 (33.2338	-119.4559) which is San Nicolas Island off of CA. AFAIK, San Nicolas is used as a weapons munition testing site and has neither an official ZipCode nor a 'City'. 

For the sake of this POC, we'll be dropping all rows with null values.

```python
df_2020_cleaned = df_2020_C_Z_dropped.reset_index()
df_2020_cleaned = df_2020_cleaned.drop(columns = ['index'])
df_2020_cleaned
```

```python
df_2020_cleaned.to_csv('weather_2020.csv', index = True)
```

#### Get data from Jan 2020 to April 2020

```python
df_jan_apr = df_2020_cleaned.loc[
            (df_2020_cleaned['StartTime(UTC)'] >= '2020-01-01') & 
            (df_2020_cleaned['StartTime(UTC)'] <= '2020-04-30')]
df_jan_apr = df_jan_apr.reset_index().drop(columns=['index'])

df_jan_apr.to_csv('weather_jan_apr_2020.csv', index = True)
```

#### Get data from Jan 2020 to August 2020

```python
df_2020_jan_aug = df_2020_cleaned.loc[
                (df_2020_cleaned['StartTime(UTC)'] >= '2020-01-01') &
                (df_2020_cleaned['StartTime(UTC)'] <= '2020-08-31')
].reset_index().drop(columns=['index'])
df_2020_jan_aug.to_csv('weather_jan_aug_2020.csv', index = True)
```
