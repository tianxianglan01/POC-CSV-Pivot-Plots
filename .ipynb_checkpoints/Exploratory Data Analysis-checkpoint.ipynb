#### Get all Unique values of Column: Type

```python
unique_types = df_2020['Type'].unique()
sorted(unique_types)
```

#### Calculate Max and Min Values of Latitude of a given DF

```python
selected_df = df_2020

max_value = selected_df.max(axis = 0)['LocationLat']
min_value = selected_df.min(axis = 0)['LocationLat']
print('max value:', max_value)
print('min value:', min_value)
```

```python
selected_df
```

#### Calculate Deciles of a given DF

```python
selected_df['Lat Decile'] = pd.qcut(selected_df['LocationLat'], 10)
```

```python
selected_df
```

#### Calculate Deciles rank of a given DF

```python
selected_df['Lat Decile Rank'] = pd.qcut(selected_df['LocationLat'], 10, labels = False)
```

```python
selected_df
```

#### Add a month Column

```python
selected_df['Month'] = pd.DatetimeIndex(selected_df['StartTime(UTC)']).month
```

```python
selected_df
```

### Data Analysis 1 Mental Model
- 12 months with 10 deciles per month on the x labels
- all event types, ['Cold', 'Fog', 'Hail', 'Precipitation', 'Rain', 'Snow', 'Storm'], on the y labels

```python
selected_columns = selected_df[['Month', 'Lat Decile Rank', 'Lat Decile', 'Type']]

selected_columns
```

```python
# code used to play around with pivot_table
analysis_incomplete = pd.pivot_table(selected_columns, 
                    columns = ['Type'],
                    index = ['Month'], 
                    aggfunc = 'count')

analysis_incomplete

# note: having a heirarchal index causes issues for getting count of type for month, and then Lat Decile Rank. 
```

```python
analysis_1 = selected_columns.pivot_table(
                columns = ['Type'],
                           index = ['Month', 'Lat Decile Rank', 'Lat Decile'],
                           aggfunc = len)
# user can decide on decile rank or decile range

analysis_1
```

#### Output to a csv

```python
analysis_1.to_csv('C:/Users/tlan4/Documents/Practice Projects/POC-CSV-Pivot-Plots/raw_and_cleaned_data/Output Csvs/types_by_latitude_by_month.csv', index = False)
```
